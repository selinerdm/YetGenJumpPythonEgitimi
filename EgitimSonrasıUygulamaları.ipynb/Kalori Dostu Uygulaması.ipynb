```python
import pandas as pd

data=pd.read_excel("D:/dir_python/calories.xlsx")

cals_splitted=[]
for i in range(0,len(data)):
    cals_splitted.append(float(data.loc[i,"Cals_per100grams"].split()[0]))
data["Cals_per100grams"]=cals_splitted
data.index=data["FoodItem"]

import time
class Kullanici():
    def __init__(self,ad,soyad,cinsiyet,yas,kilo,boy):
        self.ad=ad
        self.soyad=soyad
        self.cinsiyet=cinsiyet
        self.yas=yas
        self.kilo=kilo
        self.boy=boy
    def kullanici_kayit(self):
        with open(f"{self.ad}_{self.soyad}_kaloridostu.csv","w",encoding="utf-8") as file:
            file.writelines(f"Name,Surname,Gender,Age,Weight,Length,Basal Metabolism,Ideal Calorie,Date,Food name,Calorie Content of Food (per 100 gr)\n")
            
    def kayit_silme(self):
        import os
        os.remove(f"{self.ad}_{self.soyad}_kaloridostu.csv")

    def bazal_metabolizma(self):
        if self.cinsiyet =="male":
            global bazal_erkek
            bazal_erkek = (66.5 + (13.75*self.kilo) + (5*self.boy) - (6.77*self.yas))
            
        elif self.cinsiyet== "female":
            global bazal_kadin
            bazal_kadin= 655.1 + (9.56*self.kilo) + (1.85*self.boy) - (4.67*self.yas)
        else:
            print("\nYour basal metabolism could not be calculated.")

            
    
    def kalori_alimi(self):
        karar= int(input("""Please select your daily activity level from the options below and dial the number next to it:
        1) Less Activity
        2) Light Activity
        3) Moderate Activity
        4) High Activity"""))
        if self.cinsiyet=="male":
            if karar == 1:
                global ideal_kalori
                ideal_kalori= bazal_erkek*1.2
            elif karar == 2:
                ideal_kalori= bazal_erkek*1.375
            elif karar== 3:
                ideal_kalori= bazal_erkek*1.55
            elif karar == 4:
                ideal_kalori= bazal_erkek*1.725
        elif self.cinsiyet=="female":
            if karar == 1:
                ideal_kalori= bazal_kadin*1.2
            elif karar == 2:
                ideal_kalori= bazal_kadin*1.375
            elif karar== 3:
                ideal_kalori= bazal_kadin*1.55
            elif karar == 4:
                ideal_kalori= bazal_kadin*1.725
        else:
            print("\nYour ideal daily calories could not be calculated because your basal metabolism could not be calculated")
        
    def alim_ekleme(self):
        kalori_list=list()
        try:
            with open(f"{self.ad}_{self.soyad}_kaloridostu.csv","r",encoding="utf-8") as file:
                reader=file.read()
            while True:
                import difflib
                besin_karar = input("Enter the name of the food you want to add (each value is per 100 gr), if you do not want to add to it please press the 'q' button:")
                bulunan = difflib.get_close_matches(besin_karar,data.index,n=5,cutoff=0.4)

                if besin_karar in data["FoodItem"] :
                    from datetime import date
                    tarih=date.today()
                    tarih_str= tarih.strftime("%d/%m/%Y")

                    with open(f"{self.ad}_{self.soyad}_kaloridostu.csv","a",encoding="utf-8") as file:
                        file.writelines(f"{self.ad.capitalize()},{self.soyad.capitalize()},{self.cinsiyet.capitalize()},{self.yas},{self.kilo},{self.boy},{bazal_erkek if self.cinsiyet=='male' else bazal_kadin},{ideal_kalori},{tarih_str},{data.loc[besin_karar, 'FoodItem']},{data.loc[besin_karar,'Cals_per100grams']} \n")
                    kalori_list.append(data.loc[besin_karar,"Cals_per100grams"])

                elif besin_karar == "q":
                    toplam= sum(kalori_list)
                    print(f"\nThe total number of calories you have added:{toplam}")
                    break   
                
                else:
                    print(f"\nNo such food was found. Did you mean that: {bulunan}")
        except FileNotFoundError:
            print(f"User {self.ad.capitalize()} {self.soyad.capitalize()}  not found. To report food intake, please first create a user record for {self.ad} {self.soyad}. ")







        
                
def kalori_dostu():
    print("""Welcome to the calorie friendly app.

With its calorie friendly app, you can do the following actions:

1.Create a user record
2.Calculate the basal metabolism
3.Calculate the ideal daily calorie intake
4.Create a calorie intake record
5.View a calorie intake table by date
6.Prepare a calorie intake report
7.Delete a user

Press the 'q' button to exit the application.""")
    while True:
        islem= input("Click on the operation you want to do:")
        if islem=="1":
            ad_1=input("Enter your name:").lower()
            soyad_1= input("Enter your surname:").lower()
            cinsiyet_1= input("Enter your gender:").lower()
            yas_1= int(input("Enter your age:"))
            kilo_1=int(input("Enter your weight:"))
            boy_1= int(input("Enter your length:"))
            kaydedici=Kullanici(ad_1,soyad_1,cinsiyet_1,yas_1,kilo_1,boy_1)
            kaydedici.kullanici_kayit()
            print("\nUser record is successful.")
        elif islem=="2":
            cinsiyet_2= input("Enter your gender:").lower()
            yas_2= int(input("Enter your age:"))
            kilo_2=int(input("Enter your weight:"))
            boy_2= int(input("Enter your length:"))
            metabolizma= Kullanici("Default_Name","Default_Surname",cinsiyet_2,yas_2,kilo_2,boy_2)
            metabolizma.bazal_metabolizma()
            if cinsiyet_2=="male":
                print(f"\nYour basal metabolism:{bazal_erkek}")
            elif cinsiyet_2=="female":
                print(f"\nYour basal metabolism:{bazal_kadin}")
            else:
                print("\nPlease check the gender you entered:")
        elif islem=="3":
            cinsiyet_3= input("Enter your gender:").lower()
            yas_3= int(input("Enter your age:"))
            kilo_3=int(input("Enter your weight:"))
            boy_3= int(input("Enter your length:"))
            ideal= Kullanici("Default_Name","Default_Surname",cinsiyet_3,yas_3,kilo_3,boy_3)
            ideal.bazal_metabolizma()
            ideal.kalori_alimi()
            if cinsiyet_3=="male":
                print(f"\nYour ideal number of calories : {ideal_kalori} cal")
            elif cinsiyet_3=="female":
                print(f"\nYour ideal number of calories: {ideal_kalori} cal")
            else:
                print("\nPlease check the gender you entered:")
        elif islem=="4":
            ad_4=input("Enter your name:").lower()
            soyad_4= input("Enter your surname:").lower()
            cinsiyet_4= input("Enter your gender:").lower()
            yas_4= int(input("Enter your age:"))
            kilo_4=int(input("Enter your weight:"))
            boy_4= int(input("Enter your length:"))
            alim= Kullanici(ad_4,soyad_4,cinsiyet_4,yas_4,kilo_4,boy_4)
            alim.bazal_metabolizma()
            alim.kalori_alimi()
            alim.alim_ekleme()
        elif islem=="5":
            ad_5=input("Type the name of the user whose historical data you want to know:").lower()
            soyad_5=input("Type the surname of the user whose historical data you want to know:").lower()
            try:
                kullanici_df =pd.read_csv(f"{ad_5}_{soyad_5}_kaloridostu.csv")
                print(kullanici_df)
            except FileNotFoundError:
                print(f"\nUser {ad_5.capitalize()} {soyad_5.capitalize()} not found.\nMake sure you have entered the username correctly.If you have not already created a record for this user, create the user recor first.")
        elif islem=="6":
            try:
                ad_6=input("Type the name of the user for whom you want to prepare the daily calorie report:").lower()
                soyad_6=input("Type the surname of the user for whom you want to prepare the daily calorie report:").lower()
                kullanici_df_2 = pd.read_csv(f"{ad_6}_{soyad_6}_kaloridostu.csv")
                rapor_istek= input("Press 1 if you want to prepare a calorie report for all the days available or press 2 if you want to prepare a calorie report for a spesific day.")
                gunluk_toplam_cal=kullanici_df_2.groupby("Date")["Calorie Content of Food (per 100 gr)"].sum()
                gunluk_mean_ideal=kullanici_df_2.groupby("Date")["Ideal Calorie"].mean()
                merged_kiyas= pd.merge(gunluk_toplam_cal,gunluk_mean_ideal,on="Date")
                merged_kiyas["Date"]=merged_kiyas.index
                merged_kiyas.columns=["Total amount of calories ingested", "Ideal Calorie","Date"]
                merged_kiyas.index= list(range(0,len(list(merged_kiyas.index))))
                for i in range(0,len(list(merged_kiyas.index))):
                        rapor_sonuc=pd.Series()
                        if merged_kiyas.loc[i,"Total amount of calories ingested"] < merged_kiyas.loc[i,"Ideal Calorie"]:
                            rapor_sonuc[i]="Your calorie intake is below your ideal calorie!"
                        else:
                            rapor_sonuc[i]="You have achieved your ideal calorie intake."
                        merged_kiyas["Report Result"] = rapor_sonuc
                if rapor_istek=="1":
                    print(merged_kiyas)
                else:
                    tek_gun= input("Write the day you want to preapre a calorie report in Day/Month/Year format.")
                    merged_kiyas.index = merged_kiyas["Date"]
                    print(merged_kiyas.loc[tek_gun])
            except FileNotFoundError:
                print(f"\nUser {ad_6.capitalize()} {soyad_6.capitalize()} not found.\n.Make sure you have entered the username correctly.If you have not already created a record for this user, create the user recor first.")          
        elif islem=="7":
            ad_7=input("Enter the name of the user you want to delete:").lower()
            soyad_7=input("Enter the surname of the user you want to delete:").lower()
            silici=Kullanici(ad_7,soyad_7,"a",0,0,0)
            silici.kayit_silme()
        elif islem == "q":
            print("\nApp is closing...\nWe wish you healthy days.")
            break


kalori_dostu()




```
