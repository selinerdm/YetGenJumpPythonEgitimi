# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
class Vehicle:
    def __init__(self,model,capasity):
       self.model = model 
       self.capacity = capasity
       print("Vehicle nesnesi üretildi.")
    def intro(self):
        print(self.model,self.capacity)
class Bus(Vehicle):
    def __init__(self,model,capacity):
        super().__init__(model,capacity)
        print("Bus nesnesi çalıştı")
    def intro(self):
        print(f"Oturma kapasitesi {self.capacity}")

    
b1 = Bus("Volvo", "50")
b1.intro()
````
