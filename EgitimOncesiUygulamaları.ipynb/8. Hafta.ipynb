# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
# 3.Problemin Çözümünü Buraya Yazınız
class vehicle():
    def__init__(self,model,renk,silindir)
        self.model= model
        self.renk = renk
        self.silindir = silindir
bus = vehicle("School Volvo speed","Beyaz",5)
for bus in vehicle:
    print(f"{bus.model}: {bus.renk}, {bus.silindir}")



````
